```python
import fastf1 as ff1
from fastf1 import plotting
from matplotlib import pyplot as plt
from matplotlib.pyplot import figure
import os
```

```python
ff1.Cache.enable_cache('cache')
```

```python
session = ff1.get_session(2023, 'Monaco', 'Q')
session.load()
```

```python
session.results.columns
session.results.iloc[0:10].loc[:, ['Abbreviation', 'Q3']]
```

```python
verstappen_lap = session.laps.pick_driver('VER').pick_fastest()
alonso_lap = session.laps.pick_driver('ALO').pick_fastest()

verstappen_car = verstappen_lap.get_car_data().add_distance()
alonso_car = alonso_lap.get_car_data().add_distance()
```

```python
rbr_color = ff1.plotting.team_color('RBR')
ast_color = ff1.plotting.team_color('AST')

fig, ax = plt.subplots()
ax.plot(verstappen_car['Distance'], verstappen_car['Speed'], color=rbr_color, label='VER')
ax.plot(alonso_car['Distance'], alonso_car['Speed'], color=ast_color, label='ALO')

ax.set_xlabel('Distance in m')
ax.set_ylabel('Speed in km/h')

ax.legend()
plt.suptitle(f"Fastest Lap Comparison \n "
             f"{session.event['EventName']} {session.event.year} Qualifying")

plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(verstappen_car['Distance'], verstappen_car['Speed'], color=rbr_color, label='VER')
ax.plot(alonso_car['Distance'], alonso_car['Speed'], color=ast_color, label='ALO')
ax.legend()
plt.suptitle("Quali Speed")
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(verstappen_car['Distance'], verstappen_car['Throttle'], color=rbr_color, label='VER')
ax.plot(alonso_car['Distance'], alonso_car['Throttle'], color=ast_color, label='ALO')
ax.legend()
plt.suptitle("Quali Throttle")
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(verstappen_car['Distance'], verstappen_car['Brake'], color=rbr_color, label='VER')
ax.plot(alonso_car['Distance'], alonso_car['Brake'], color=ast_color, label='ALO')
ax.legend()
plt.suptitle("Quali Brake")
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(verstappen_car['Distance'], verstappen_car['Brake'], color='red', label='BRAKE')
ax.plot(verstappen_car['Distance'], verstappen_car['Throttle'], color='green', label='THROTTLE')
ax.legend()
plt.suptitle("VER Fastest Lap")
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(alonso_car['Distance'], alonso_car['Brake'], color='red', label='BRAKE')
ax.plot(alonso_car['Distance'], alonso_car['Throttle'], color='green', label='THROTTLE')
ax.legend()
plt.suptitle("ALO Fastest Lap")
plt.show()
```
